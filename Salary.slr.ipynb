```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("Salary_Data.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(15,7),facecolor="lightblue")
plt.scatter(df.YearsExperience,df.Salary)
plt.xlabel('Yearexperience')
plt.ylabel('Salary')
plt.show()
```

```python
df.corr()
```

```python
model= smf.ols("YearsExperience~Salary",data=df).fit()
```

```python
model.summary()
```

```python
fm = smf.ols("YearsExperience~np.log(Salary)",data=df).fit()#log
```

```python
fm.summary()
```

```python
fm = smf.ols("np.log(YearsExperience)~Salary",data=df).fit() #log
fm.summary()
```

```python

df["salary1"]= df.Salary**2
df
```

```python
mq = smf.ols("np.log(YearsExperience)~Salary+salary1",data=df).fit()
```
